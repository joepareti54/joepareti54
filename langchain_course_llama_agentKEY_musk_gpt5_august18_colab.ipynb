```python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig, AutoTokenizer
from langchain_huggingface import ChatHuggingFace
from langchain.agents import initialize_agent, AgentType, load_tools
#from langchain_community.tools import PythonREPLTool
from langchain_experimental.tools import PythonREPLTool
from huggingface_hub import login
import torch
```

```python
import os
from google.colab import userdata
# Get API key from Colab secrets
serpapi_api_key = userdata.get('SERPAPI_API_KEY')
os.environ['SERPAPI_API_KEY'] = serpapi_api_key
hf_token = userdata.get('HUGGINGFACE_TOKEN')
login(token=hf_token)
```

```python
bnb_config = BitsAndBytesConfig(
load_in_4bit=True,
bnb_4bit_quant_type="nf4",
bnb_4bit_compute_dtype=torch.float16,
bnb_4bit_use_double_quant=True
)
llm_name = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(llm_name)
llm = AutoModelForCausalLM.from_pretrained(
    llm_name,
    device_map="auto",
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True
)
inputs = tokenizer("I want to open a restaurant for Indian food. Suggest a fancy name for this.", return_tensors="pt").to("cuda")
outputs = llm.generate(**inputs, max_new_tokens=100)
#name = llm("I want to open a restaurant for Indian food. Suggest a fency name for this.")
#print(name)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
pipe = pipeline(
    "text-generation",
    model=llm,  # this is your loaded LlamaForCausalLM
    tokenizer=tokenizer,
    temperature=0.1,
    max_new_tokens=256
)
langchain_llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.agents import AgentType, initialize_agent, load_tools
```

```python
tools = [PythonREPLTool()] + load_tools(["serpapi"])
agent = initialize_agent(
    tools=tools,
    llm=langchain_llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
#    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    max_iterations=3,
    handle_parsing_errors=(
        "If you mixed Final Answer with Action, output ONLY one: "
        "either the next Action + Action Input OR a Final Answer."
    )
)
agent.run("What is Elon Musk's exact date of birth ? Calculate his exact age on May 5 2011 ")
```
